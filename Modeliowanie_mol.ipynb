```python
#An expamle of what I am  going to gain
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
 
# Create a dataset
my_count=["France","Australia","Japan","USA","Germany","Congo","China","England","Spain","Greece","Marocco","South Africa","Indonesia","Peru","Chili","Brazil"]
df = pd.DataFrame({
"country":np.repeat(my_count, 10),
"years":list(range(2000, 2010)) * 16,
"value":np.random.rand(160)
})
 
# Create a grid : initialize it
g = sns.FacetGrid(df, col='country', hue='country', col_wrap=4, )

# Add the line over the area with the plot function
g = g.map(plt.scatter,'years', 'value', marker = '*')
 
# Fill the area with fill_between
g = g.map(plt.fill_between, 'years', 'value', alpha=0.2).set_titles("{col_name} country")
 
# Control the title of each facet
g = g.set_titles("{col_name}")
 
# Add a title for the whole plot
plt.subplots_adjust(top=0.92)
g = g.fig.suptitle('Evolution of the value of stuff in 16 countries')

# Show the graph
plt.show()
```

```python
from mpl_toolkits.axisartist.axislines import AxesZero
#from matplotlib.patches import Rectangle
def set_size(w,h, ax=None):
    """ w, h: width, height in inches """
    if not ax: ax=plt.gca()
    l = ax.figure.subplotpars.left
    r = ax.figure.subplotpars.right
    t = ax.figure.subplotpars.top
    b = ax.figure.subplotpars.bottom
    figw = float(w)/(r-l)
    figh = float(h)/(t-b)
    ax.figure.set_size_inches(figw, figh)
x = mk.Sample
y = mk.Mark
y1 = mk.e_coli_conc_2004_2006
y2 = mk.e_coli_conc_2007_2008
y3 = mk.e_coli_conc_2009_2010
y4 = mk.e_coli_conc_2016_2020
fig, ax = plt.subplots()
x_coordinates = [0, 15]
y_coordinates = [500, 500]
x_coordinates2 = [0, 15]
y_coordinates2 = [250, 250]
size = 1100
#ax.add_patch(Rectangle((2, 2), 1,))
ax.scatter(x,y1,color ='red',marker = '^',s = size)
ax.scatter(x,y2,color ='orange',marker = 'h',s = size)
ax.scatter(x,y3,color ='olive',marker = 'o',s = size)
ax.scatter(x,y4,color ='green',marker = 'X',s = size)
ax.set_label(['E.coli con. 2004/2006','E.coli conc. 2007/2008','E.coli conc. 2009/2010','E.coli conc. 2016/2020'])
#ax.legend(['E.coli conc. 2004/2006','E.coli conc. 2007/2008','E.coli conc. 2009/2010','E.coli conc. 2016/2020'],loc='center left',fontsize=25)
fontsize = 40
plt.tick_params(axis='both', which='major', labelsize= 28)
ax.set_ylabel('E.coli concentration [CFU(100mL^-1)]',fontsize=fontsize)
#ax.plot(x_coordinates,y_coordinates, color = 'r', )
#ax.plot(x_coordinates2,y_coordinates2, color = 'y')
ax.axhspan(0, 250, facecolor='green', alpha=0.05)#set the arrea for danger level
ax.axhspan(250, 500, facecolor='yellow', alpha=0.05)# set the arrea for semi danger level
ax.axhspan(500, 3200, facecolor='red', alpha=0.05)# set the arrea for sefa level 
#ax.set_xlabel('Points',fontsize=fontsize);
plt.xlim(-0.15,14.2)
plt.ylim(-5,3200)
set_size(30,20)
plt.style.use('seaborn-talk')
plt.savefig('fig_water_qual(modef).png', dpi = 300, bbox_inches='tight')
```

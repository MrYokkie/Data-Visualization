```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "ID":["Waste landfill siol","Referance soil","Control sand"],
    "Sample":["Waste landfill siol","Referance soil","Control sand"],
    "Sorghum sacharatum":[657,1763,1656],
    "Sinapis alba":[819,1253,997]
})
```

```python
df
```

```python

f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=df, x="Sample", y="Sorghum sacharatum", hue="Sample", ax=axs[0])
sns.histplot(data=df, x="Sorghum sacharatum", hue="Sample", shrink=.8, alpha=.8, legend=False)
f.tight_layout()
```

```python
df1 = pd.DataFrame({
    "ID":["Waste_landfill_soil","Referance_soil","Control_sand","Waste_landfill_soil","Referance_soil","Control_sand"],
    "Sample":["Waste_landfill_soil","Waste_landfill_soil","Referance_soil","Referance_soil","Control_sand","Control_sand"],
    "Name_of_tasted_plant":["Sorghum sacharatum","Sinapis alba","Sorghum sacharatum","Sinapis alba","Sorghum sacharatum","Sinapis alba"],
    "Length_of_roots":[657,819,1763,1253,1655,997]
})
```

```python
sns.histplot(data=df1, x="Sample", y="Length_of_roots", hue="Name_of_tasted_plant")
```

```python
g = sns.catplot(
    data=df1, kind="bar",
    x="Sample", y="Length_of_roots", hue="Name_of_tasted_plant",
    ci="sd", palette="crest", alpha=.6, height=6, legend= False, aspect= 13/12
)
g.despine(left=True)
g.set_axis_labels("", "Length of roots [mm]")
plt.legend(title='Name_of tasted plant', loc='lower right')
plt.savefig('Phyto_wyk.png', dpi =300, bbox_inches='tight')
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=df1, kind="bar",
    x="ID", y="Length_of_roots", hue="Name_of_tasted_plant",
    ci="sd", palette='crest', alpha=.8, height=6,legend= False, aspect= 14/13,
)
g.despine(left=True)
g.set_axis_labels("", "Length of roots (mm)", fontsize = 15)
plt.legend(title='Name_of_tasted_plant', loc='lower right', fontsize = 10)
plt.savefig('Phyto_wyk.png', dpi =300, bbox_inches='tight')
```

```python
mic = pd.DataFrame({
    "Hormesis_effect_(+/-)": ['+','+','-','-','+','+','-',None],
    "Sample":['Waste_landfill_soil','Waste_landfill_soil','Referance_soil','Referance_soil','Waste_landfill_soil','Waste_landfill_soil','Referance_soil','Referance_soil'],
    "Time_of_contact_with_bacteria_(min)":[15,30,15,30,15,30,15,30],
    "Toxicity effect_(EC50_in_%)":[-37,-37,172,223,-19,-27,173,None]
})
```

```python
mic
```

```python
g = sns.catplot(x="Sample", y="Toxicity effect_(EC50_in_%)", hue="Time_of_contact_with_bacteria_(min)",
                capsize=.7, palette='Dark2', height=6, aspect=.75,
                kind="point", data=mic)
g.despine(left=True)
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=mic, kind="bar",
    x="Sample", y="Toxicity effect_(EC50_in_%)", hue="Time_of_contact_with_bacteria_(min)",
    ci="sd", palette='icefire', alpha= 1, height=8, legend= False
)
g.despine(left=True)
g.set_axis_labels("Sample", "Toxicity effect_(EC50_in_%)",fontsize =17)
plt.legend(title="Time_of_contact_with_bacteria_(min)", loc='lower right', fontsize = 10)
plt.savefig('microtox.png',pdi = 300, )
```

```python
sns.set_context("notebook", font_scale=1)
g = sns.catplot(
    data=mic, kind="bar",
    x="Sample", y="Toxicity effect_(EC50_in_%)", hue="Time_of_contact_with_bacteria_(min)",
    ci="sd", palette='dark', alpha=.6, height=6,legend = False, aspect= 15/12
)
g.despine(left=True)
g.set_axis_labels("Sample", "Toxicity effect_(EC50 in %)")
plt.legend(title="Time of contact with bacteria(min)", loc='lower right')
plt.savefig('Micro_res.png', dpi = 300, bbox_inches='tight')
```
